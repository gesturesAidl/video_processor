```python
import pandas as pd
import torch
from torch.utils.data import Dataset

class JesterDatasetOneStream(Dataset):
  
  def __init__(self,features_dict,data_path,labels_path): 
    super().__init__()
    self.features_dict = features_dict
    self.labels = pd.read_csv(labels_path)
    self.data = pd.read_csv(data_path)
  
  def __len__(self):
    return(len(self.data))
  
  def __getitem__(self,idx):
    item = self.data.iloc[idx]
    label = self.labels.loc[self.labels['Actions'] == item[1]].index[0]
    feature = torch.tensor(self.features_dict[item[0]]).view(-1)
    return feature,label

class JesterDatasetTwoStream(Dataset):
  
  def __init__(self,features_dict,flow_dict,data_path,labels_path): 
    super().__init__()
    self.features_dict = features_dict
    self.flow_dict = flow_dict
    self.labels = pd.read_csv(labels_path)
    self.data = pd.read_csv(data_path)
  
  def __len__(self):
    return(len(self.data))
  
  def __getitem__(self,idx):
    item = self.data.iloc[idx]
    label = self.labels.loc[self.labels['Actions'] == item[1]].index[0]
    rgb_feature = torch.tensor(self.features_dict[str(item[0])]).view(-1)
    flow_feature = torch.tensor(self.flow_dict[str(item[0])]).view(-1)
    
    return rgb_feature,flow_feature, label
```
